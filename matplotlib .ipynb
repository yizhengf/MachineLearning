## matplotlib 基础

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x=np.linspace(0,10,100)
```

```python
x
```

```python
y=np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
cosy=np.cos(x)
```

```python
 siny=y.copy()
```

```python
plt.plot(x,siny,'--',color="red")
plt.plot(x,cosy)
plt.show()
```

```python
plt.plot(x,siny,'--',color="red")
plt.plot(x,cosy)
plt.xlim(-5,15)
plt.ylim(0,2)
plt.show()
```

```python
plt.plot(x,siny,'--',color="red")
plt.plot(x,cosy)
plt.axis([-1,11,-2,2])
plt.show()
```

```python
plt.plot(x,siny,'--',color="red")
plt.plot(x,cosy)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
plt.plot(x,siny,'--',color="red",label="sin(X)")
plt.plot(x,cosy, label="cos(X)")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Sample")
plt.show()
```

```python
plt.scatter(x,siny)
plt.scatter(x,cosy)
```

```python
x=np.random.normal(0,1,10000)
y=np.random.normal(0,1,10000)
plt.scatter(x,y, alpha=0.1)
plt.show()
```

## 读取数据和简单的数据探索

```python
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
iris.keys()
```

```python
print(iris.DESCR)
```

```python
iris.data.shape
```

```python
iris.feature_names
```

```python
iris.target.shape # 1 dimension
```

```python
iris.target_names
```

```python
X=iris.data[:,:2]
X.shape
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
y=iris.target
```

```python
plt.scatter(X[y==0,0],X[y==0,1], color="red",marker="o")
plt.scatter(X[y==1,0],X[y==1,1], color="blue",marker="+")
plt.scatter(X[y==2,0],X[y==2,1], color="green", marker="x")
plt.show()
```

```python
X=iris.data[:,2:] #change to another two features 
```

```python
plt.scatter(X[y==0,0],X[y==0,1], color="red",marker="o")
plt.scatter(X[y==1,0],X[y==1,1], color="blue",marker="+")
plt.scatter(X[y==2,0],X[y==2,1], color="green", marker="x")
plt.show()
```
